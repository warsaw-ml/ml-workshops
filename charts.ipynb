```python
import matplotlib.pyplot as plt

# Data for AP
classes_ap = ['person_normal', 'person_topview']
ap_values = [51.183, 47.564]

# Data for AR
classes_ar = ['person_normal', 'person_topview']
ar_values = [62.137, 56.351]

# Set the width of each bar
bar_width = 0.25


# Set the positions of the bars on the x-axis
r1 = range(len(classes_ap))
r2 = [x + bar_width for x in r1]

# Create the bar chart
fig, ax = plt.subplots(figsize=(5, 6))

# Plot the bars for AP and AR
ap_bars = ax.bar(r1, ap_values, color='b', width=bar_width, edgecolor='white', label='AP')
ar_bars = ax.bar(r2, ar_values, color='g', width=bar_width, edgecolor='white', label='AR')

# Add labels and title
ax.set_xlabel('Classes')
# ax.set_ylabel('Values')
ax.set_title('AP and AR per Class\n Model finetuned on top view + normal view')
ax.set_xticks([r + bar_width/2 for r in range(len(classes_ap))])
ax.set_xticklabels(classes_ap)

ax.set_ylim(top=100)

# Add a legend
ax.legend()

# Add values on top of each bar
for bar in ap_bars:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height,
            f'{height:.3f}', ha='center', va='bottom')

for bar in ar_bars:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height,
            f'{height:.3f}', ha='center', va='bottom')

# Display the chart
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for original YOLOX model
original_classes = ['person']
original_ap_values = [34.221]
original_ar_values = [71.369]

# Set the width of each bar
bar_width = 0.25

# Create the bar chart
fig, ax = plt.subplots(figsize=(3, 6))

# Plot the bars for the original YOLOX model
x = np.arange(len(original_classes))
ax.bar(x - bar_width/2, original_ap_values, width=bar_width, color='b', label='AP')
ax.bar(x + bar_width/2 + 0.05, original_ar_values, width=bar_width, color='g', label='AR')

# Add labels and title for the original YOLOX model
ax.set_xlabel('Person class')
# ax.set_ylabel('Values')
ax.set_title('Original YOLOX model tested on sample of person class from COCO dataset')
ax.set_xticks(x)
ax.set_xticklabels([''])
ax.set_ylim(top=100)
ax.legend()

# Add values on top of each bar for the original YOLOX model
for i, v in enumerate(original_ap_values):
    ax.text(x - bar_width/2, v + 1, f'{v:.3f}', ha='center', va='bottom')
for i, v in enumerate(original_ar_values):
    ax.text(x + bar_width/2, v + 1, f'{v:.3f}', ha='center', va='bottom')

# Adjust the spacing
plt.tight_layout()

# Display the chart
plt.show()
```
